## The Order of an ODE
**The order of an ODE**: the number of the highest derivative

```sympy.solvers.deutils.ode_order(expr, func)```

Returns the order of a given differential equation with respect to func [2].

Find the order of the equation.
* (a) $\dfrac{d^2y}{dx^2}+2\dfrac{dy}{dx}\,\dfrac{d^3y}{dx^3}+x=0$
* (b) $y''-3y'+2y=x^7$
* (c) $y'-y^7=0$
* (d) $y''y-(y')^2=2$

````python
from sympy import *
init_printing(use_unicode=True)
#
from sympy import Function, dsolve, Eq, Derivative, sin, cos, symbols
from sympy.solvers.deutils import ode_order
from NA_Notes import Header, Line

x, y, z, c, a = symbols('x y z c a')
y = Function('y')

# (a)
Header('(a)')
PDE_Order = ode_order(Derivative(y(x), x, x)+Derivative(y(x), x)*Derivative(y(x), x, x,x)+x, y(x))
print('The order of the ODE = %i' % PDE_Order)
Line()
del PDE_Order
````

````python
# (b)
Header('(b)')
PDE_Order = ode_order(Derivative(y(x), x, x)-3*Derivative(y(x), x)+2*y(x)-x**7, y(x))
print('The order of the ODE = %i' % PDE_Order)
Line()
del PDE_Order
````

````python
# (c)
Header('(a)')
PDE_Order = ode_order(Derivative(y(x), x)-y(x)**7, y(x))
print('The order of the ODE = %i' % PDE_Order)
Line()
del PDE_Order
````

````python
# (d)
Header('(d)')
PDE_Order = ode_order(Derivative(y(x), x, x)*y(x)-Derivative(y(x), x)**2-2, y(x))
print('The order of the ODE = %i' % PDE_Order)
Line()
del PDE_Order
````

* (a) $y=ce^{2x};     \quad y'=2y$

```sympy.solvers.ode.dsolve(eq, func=None, hint='default', simplify=True, ics=None, xi=None, eta=None, x0=0, n=6, **kwargs)```

Solves any (supported) kind of ordinary differential equation and system of ordinary differential equations [3].

Find all solutions of the equation.

* (a) $y'=-x$
* (b) $y'=-x \sin x$
* (c) $y''=2xe^x$

````python
# (a)
Header('(a)')
Sol = dsolve(Derivative(y(x)) +x, y(x))
display(Sol)
Line()
del Sol
````

````python
# (b)
Header('(b)')
Sol = dsolve(Derivative(y(x)) +x*sin(x), y(x))
display(Sol)
Line()
del Sol
````

````python
# (b)
Header('(a)')
Sol = dsolve(Derivative(y(x), x, x) -2*x*exp(x), y(x))
display(Sol)
Line()
del Sol
````

</div>

In doing so, we simply plug the solution in the ODE and see whether both sides are equal or not.

Verify that the  function is a solution of the initial value problem.
* (a) $y=x\cos x;     \quad y'=\cos x-y\tan x, \quad y(\pi/4)=\dfrac{\pi}{4\sqrt{2}}$
* (b) $y=\dfrac{1+2\ln x}{x^2}+\dfrac{1}{2};     \quad y'=\dfrac{x^2-2x^2y+2}{x^3},   \quad y(1)=\dfrac{3}{2}$

````python
# (a)
Header('(a)')
y= x*cos(x)
Sol = expand(diff(y,x)-cos(x)+y*tan(x))
display(Sol)
#
from colorama import Fore, Style
if Sol == 0:
    print(Fore.GREEN + Style.NORMAL + 'y is a solution' + Style.RESET_ALL)
else:
    print(Fore.RED + Style.NORMAL + 'y is not a solutiont' + Style.RESET_ALL)
Line()
del Sol
````

````python
# (b)
Header('(b)')
y= (1+2*ln(x))/(x**2)+(1/2)
Sol = expand(diff(y,x)-((x**2-2*x**2*y+2)/(x**3)))
display(Sol)
#
if Sol == 0:
    print(Fore.GREEN + Style.NORMAL + 'y is a solution' + Style.RESET_ALL)
else:
    print(Fore.RED + Style.NORMAL + 'y is not a solutiont' + Style.RESET_ALL)
Line()
del Sol
````
