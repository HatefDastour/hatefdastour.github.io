## The elementary theory of initial-value problem

In this section, we will look at many approaches for approximating the solution of a (well-posed) initial-value problem (IVP) using grid points. The first-order Ordinary Differential Equation (ODE) is written as,

\begin{equation}
\frac{dy}{dt} = f(t,y).
\end{equation}
If $y(a) = \alpha$  is known, then the following first-order ODE equation is an IVP:
\begin{equation}
\begin{cases}
\dfrac{dy}{dt} = f(t,y),& a \leq t \leq b,\\
y(a) = \alpha.&
\end{cases}
\end{equation}

The following definition is **Definition 5.1** from {cite:ps}`burden2005numerical`.

`````{admonition} Limit of a function
:class: hint
Let $f(t,y)$ be a function, where $t$ is a independent variable and $y$ is a dependent variable, defined on a domain $D\subset \mathbb{R}^{2}$. In addition, if
\begin{equation}\label{Lipschitz condition}
| f(t, y_{1}) - f(t, y_{2}) | \leq L|y_{1} - y_{2}|,\quad (t, y_{1}), (t, y_{2}) \in D,
\end{equation}
then it is said that $f$ satisfies the **Lipschitz condition** in the
variable $y$, and $L$ is known as the **Lipschitz constant**.
`````

<font color='Blue'><b>Example</b></font>: $f (t, y) = t y^2$ does not satisfy any Lipschitz condition on the region 
$D = \{(t, y)~|~0 \leq t \leq T \}$. However, $f (t, y) = t y^2$ satisfies Lipschitz condition on the
region \linebreak $D = \{(t, y)~|~0 \leq t \leq T,~-Y \leq y \leq Y \}$ where $T$ and $Y$ are some real numbers {cite:ps}`Gu_LectureNotes`.

The following example is based on an example from {cite:ps}`burden2005numerical`.

<font color='Blue'><b>Example</b></font>:
Demonstrate that $f (t,~y) = t| y|$ satisfies a Lipschitz condition on the interval $D = \{(t,~y)~|~1 \leq
t \leq 2\text{ and }- 3 \leq y \leq 4\}$.

<font color='Green'><b>Solution</b></font>:
For any pair of points $(t,~y_{1})$ and $(t,~y_{2})$ from $D$, we have,
\begin{align*}
|f(t,~y_{1}) - f (t,~y_{2})| = |t|y_{1}| - t| y_{2}|| = |t(|y_{1}| - |y_{2}|)|
 = |t| {\Big|}|y_{1}| - |y_{2}| {\Big|}
\leq  2|y_{1} - y_{2}|.
\end{align*}
The above shows that $f$ satisfies a Lipschitz condition on D in variable $y$ and with Lipschitz constant 2. Therefore, for this problem, the smallest Lipschitz constant would be $L = 2$, as, for example,
\vspace{-0.5cm}
\begin{equation*}
|f (2, 1) - f (2, 0)| = |2 - 0| = 2|1 - 0|.
\end{equation*}


```{admonition} **Triangle inequality**
For any real numbers $\alpha$ and $\beta$, we have {cite:ps}`rudin1976principles`,
\vspace{-0.5cm}
\begin{equation}
|\alpha - \beta| \geq {\biggl |}|\alpha|-|\beta|{\biggr |}
\end{equation}
```

The following definition is **Definition 5.2** from {cite:ps}`burden2005numerical`.

`````{admonition} Limit of a function
:class: hint
Set $D\subset \mathbb{R}^{2}$ is called **convex** if any two points of
this domain can be connected using a line while all points on this line
stay in the set. In other words, for any two points
$(t_{1}, y_{1}), (t_{2}, y_{2}) \in D$,
\begin{equation}
((1 - \lambda)t_{1} + \lambda t_{2}, (1 - \lambda)y_{1} + \lambda y_{2}) \in D, \quad \lambda \in [0,1].
\end{equation}
`````

The following theorem is **Theorem 5.2** from {cite:ps}`burden2005numerical`.

```{admonition} Theorem
Assume that $D\subset \mathbb{R}^{2}$ is a convex set and $f(t,y)$ is defined on this set. If a constant $L>0$ exists such that
\begin{equation}
\left| \frac{\partial f}{\partial y} (t, y) \right| \leq L,\quad (t, y) \in D,
\end{equation}
then $f$ satisfies the Lipschitz condition in the variable $y$ and $L$ is the Lipschitz constant.
```

The following definition is **Theorem 5.4** from {cite:ps}`burden2005numerical`.

```{admonition} Theorem
Assume that domain $D = \{ (t, y):~a\leq t \leq b,~-\infty<y< \infty\}$ and $f\in C(D)$. If $f$ satisfies a Lipschitz condition on $D$ in the variable $y$, then the IVP \cite{burden2005numerical},
\begin{equation}
\begin{cases}
\dfrac{dy}{dt} = f(t,y),& a \leq t \leq b,\\ 
y(a) = \alpha,&
\end{cases}
\end{equation}
has a unique solution $y(t)$ when $a\leq t \leq b$.
```


The following definition is **Definition 5.5** from {cite:ps}`burden2005numerical`.

`````{admonition} 
:class: hint

The following IVP

\begin{equation}\label{wellpossed_eq01}
\begin{cases}
\dfrac{dy}{dt} = f(t,y),& a \leq t \leq b,
\\ \vspace{-0.6cm} \\
y(a) = \alpha,&
\end{cases}
\end{equation}

is a **well-posed** problem if {cite:ps}`burden2005numerical`

* a. it has a solution and this solution is **unique**,

* b. constants $\varepsilon_{0} >0$ and $k>0$ exists such that for any arbitrary value of $\varepsilon > 0$ with $\varepsilon_{0} > \varepsilon > 0$, whenever $\delta(t)$ is continuous with $|\delta(t)|<\varepsilon$ for all values of $t \in [a,b]$, and when $|\delta_{0}| < \varepsilon$, the IVP

\begin{equation}\label{wellpossed_eq02}
\begin{cases}
\dfrac{dz}{dt} = f(t,z) + \delta(t),& a \leq t \leq b, \\
z(a) = \alpha + \delta_{0},&
\end{cases}
\end{equation}

has a unique solution $z(t)$ that satisfies
\begin{equation}
|z(t) - y(t) | < k \varepsilon, \quad t \in[a,b]
\end{equation}

`````

`````{admonition} Remark
:class: important

Roughly speaking, an IVP is said to be a **well-posed** problem if it
has a unique solution whose value varies only negligibly if initial
conditions change by a small value.

`````

The following theorem is **Theorem 5.6** from {cite:ps}`burden2005numerical.

`````{admonition} Theorem

Assume that domain $D = \{ (t, y):~a\leq t \leq b,~-\infty<y< \infty\}$ and $f\in C(D)$. If $f$ satisfies a Lipschitz condition on $D$ in the variable $y$, then the IVP {cite:ps}`burden2005numerical`,
\begin{equation}
\begin{cases}
\dfrac{dy}{dt} = f(t,y),& a \leq t \leq b,\\ y(a) = \alpha,&
\end{cases}
\end{equation}
is well-posed.

`````
