## Model Evaluation: R-squared Statistic and Correlation Matrix
### R-squared Statistic
The R-squared statistic, denoted as $R^2$, is a measure of the proportion of the total variation in the dependent variable ($Y$) that is explained by the independent variable(s) ($X$) in a linear regression model. It ranges from 0 to 1, with 1 indicating that the model perfectly explains the variability in the dependent variable, and 0 indicating that the model does not explain any of the variability.

The formula to calculate $R^2$ is:

\begin{align}
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
\end{align}

where:
- $n$ is the number of data points in the dataset.
- $y_i$ is the actual value of the dependent variable for the $i$th data point.
- $\hat{y}_i$ is the predicted value of the dependent variable for the $i$th data point based on the regression model.
- $\bar{y}$ is the mean value of the dependent variable.

### Correlation Matrix
The correlation matrix is a square matrix that displays the correlations between different variables (parameters) in a given dataset. Each element in the matrix represents the correlation coefficient between two variables. The correlation coefficient measures the strength and direction of the linear relationship between two variables. It ranges from -1 to +1, with -1 indicating a perfect negative linear relationship, +1 indicating a perfect positive linear relationship, and 0 indicating no linear relationship.

The correlation coefficient between two variables $X$ and $Y$ can be calculated using the formula:

\begin{align}
\text{Cor}(X, Y) = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^{n} (y_i - \bar{y})^2}}
\end{align}

where:
- $n$ is the number of data points in the dataset.
- $x_i$ and $y_i$ are the values of the $X$ and $Y$ variables, respectively, for the $i$th data point.
- $\bar{x}$ and $\bar{y}$ are the mean values of $X$ and $Y$ variables, respectively.

The correlation matrix provides a convenient way to visualize the relationships between multiple variables in a dataset. It helps identify patterns of association and potential multicollinearity (high correlation between predictor variables) in regression models. A positive correlation coefficient indicates that the variables tend to increase or decrease together, while a negative correlation coefficient indicates an inverse relationship. A correlation coefficient close to 0 suggests little or no linear relationship between the variables.

## An Introduction to Linear Regression:

Linear regression is a statistical method used to model the relationship between a dependent variable (response variable) and one or more independent variables (predictors). It assumes a linear relationship between the variables, meaning that the relationship can be approximated by a straight line.

The basic linear regression model can be expressed as follows:

\begin{equation}
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \ldots + \beta_p X_p + \varepsilon,
\end{equation}

where:
- $Y$ is the dependent variable (response).
- $X_1, X_2, \ldots, X_p$ are the independent variables (predictors).
- $\beta_0, \beta_1, \ldots, \beta_p$ are the coefficients, also known as regression coefficients or model parameters.
- $\varepsilon$ is the error term, representing the unexplained variability in the model.

The goal of linear regression is to estimate the coefficients $\beta_0, \beta_1, \ldots, \beta_p$ in such a way that the model best fits the observed data. The most commonly used method to estimate these coefficients is the least-squares approach, which minimizes the sum of squared residuals {cite:p}`james2023introduction`:

\begin{equation}
\text{RSS} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2,
\end{equation}

where $y_i$ is the observed value of the dependent variable, $\hat{y}_i$ is the predicted value of the dependent variable based on the model, and $n$ is the number of data points.

To estimate the coefficients, we use a training dataset with known values of $Y$ and $X$. The goal is to find the values of $\beta_0, \beta_1, \ldots, \beta_p$ that minimize the RSS. This is typically done using numerical optimization techniques.

Once the coefficients are estimated, we can use the linear regression model to make predictions for new data points. Given the values of the independent variables $X_1, X_2, \ldots, X_p$, we can calculate the predicted value of the dependent variable $Y$ using the equation {cite:p}`james2023introduction`:

\begin{equation}
\hat{Y} = \hat{\beta_0} + \hat{\beta_1} X_1 + \hat{\beta_2} X_2 + \ldots + \hat{\beta_p} X_p.
\end{equation}

Linear regression can be used for both simple linear regression (when there is only one predictor) and multiple linear regression (when there are multiple predictors). It is widely used in various fields, including statistics, economics, social sciences, and machine learning, for tasks such as prediction, forecasting, and understanding the relationship between variables.

### Assumptions of Linear Regression:
Linear regression makes several assumptions to ensure the validity of the model and the accuracy of the estimates. Some of the key assumptions include {cite:p}`james2023introduction`:
1. Linearity: The relationship between the dependent and independent variables is linear.
2. Independence: The error terms ($\varepsilon$) are independent of each other.
3. Homoscedasticity: The variance of the error terms is constant across all levels of the predictors.
4. Normality: The error terms follow a normal distribution with mean zero.
5. No perfect multicollinearity: The predictors are not perfectly correlated with each other.

If these assumptions are violated, the accuracy and reliability of the linear regression model may be compromised. Therefore, it is essential to check these assumptions before interpreting the results of the model. Various diagnostic tests and plots, such as residual plots, can be used to assess the validity of these assumptions.

### Simple Linear Regression

Simple linear regression is a statistical technique that models the relationship between two variables: a dependent variable (also known as the response variable) and a single independent variable (also known as the predictor). It assumes a linear relationship between the variables, which means that the relationship can be approximated by a straight line.

The basic form of simple linear regression can be expressed as follows {cite:p}`james2023introduction`:

\begin{equation}
Y = \beta_0 + \beta_1 X + \varepsilon,
\end{equation}

where:
- $Y$ is the dependent variable (response).
- $X$ is the independent variable (predictor).
- $\beta_0$ is the intercept, representing the value of $Y$ when $X$ is zero.
- $\beta_1$ is the slope, representing the change in $Y$ for a one-unit change in $X$.
- $\varepsilon$ is the error term, representing the unexplained variability in the model.

The goal of simple linear regression is to estimate the coefficients $\beta_0$ and $\beta_1$ in such a way that the line best fits the observed data. The most commonly used method to estimate these coefficients is the least-squares approach, which minimizes the sum of squared residuals:

\begin{equation}\text{RSS} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2, \end{equation}

where:
- $y_i$ is the observed value of the dependent variable ($Y$) for the $i$th data point.
- $\hat{y}_i$ is the predicted value of the dependent variable ($Y$) for the $i$th data point based on the model.
- $n$ is the number of data points.

Once the coefficients $\beta_0$ and $\beta_1$ are estimated, the simple linear regression model can be represented as:

\begin{equation}\hat{Y} = \hat{\beta_0} + \hat{\beta_1} X. \end{equation}

This line represents the best linear fit to the data, and it is used to make predictions for new data points. Given the value of the independent variable ($X$) for a new data point, we can use the equation to calculate the predicted value of the dependent variable ($Y$).

Simple linear regression is a widely used technique in various fields, including economics, finance, social sciences, and engineering. It is particularly useful when there is a clear linear relationship between the two variables, allowing us to make predictions or understand the effect of the independent variable on the dependent variable {cite:p}`james2023introduction`.

However, it is essential to interpret the results of a simple linear regression analysis carefully. Factors like the goodness of fit (e.g., R-squared), statistical significance of the coefficients, and the validity of the assumptions (e.g., linearity, independence of errors, normality of errors) should be considered to draw meaningful conclusions from the model. Additionally, simple linear regression may not be appropriate if the relationship between the variables is non-linear or if there are multiple predictors that influence the dependent variable. In such cases, multiple linear regression or other more complex regression models may be more suitable {cite:p}`james2023introduction`.

In this section, we will work with the Boston Housing dataset using the `sklearn` (Scikit-learn) library, which is a popular Python library for machine learning and data science. The Boston Housing dataset is a widely used dataset in regression analysis, and it contains information about various features related to housing prices in different neighborhoods in Boston.

We only use `AGE`, `LSTAT` and `MEDV` from this dataset.

| Variable |                               Description                               |
|:--------:|:-----------------------------------------------------------------------:|
|    AGE   |          proportion of owner-occupied units built prior to 1940         |
|   LSTAT  |                     \% lower status of the population                   |
|   MEDV   |             Median value of owner-occupied homes in \$1000's            |

```python
import numpy as np
import pandas as pd

_url = "http://lib.stat.cmu.edu/datasets/boston"
columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

Boston = pd.read_csv(filepath_or_buffer= _url, delim_whitespace=True, skiprows=21,
                 header=None)

#Flatten all the values into a single long list and remove the nulls
values_w_nulls = Boston.values.flatten()
all_values = values_w_nulls[~np.isnan(values_w_nulls)]

#Reshape the values to have 14 columns and make a new df out of them
Boston = pd.DataFrame(data = all_values.reshape(-1, len(columns)),
                  columns = columns)
# Display the first 10 rows
display(Boston.head(10))
```

To start, we will fit a simple linear regression model using the `sm.OLS()` function from `statsmodels` library. In this model, our response variable will be `medv`, and `lstat` will be the single predictor.

```python
import statsmodels.api as sm

# Create the model matrix manually
X = Boston['LSTAT']  # Predictor variable (lstat)
X = sm.add_constant(X)  # Add an intercept term

# Response variable (medv)
y = Boston['MEDV']

# Fit the simple linear regression model
Results = sm.OLS(y, X).fit()

# Print the model summary
print(Results.summary())
```

```python
# Fancy dataframe for OLS Reression results
def Reg_Result(Inp):
    Temp = pd.read_html(Inp.summary().tables[1].as_html(), header=0, index_col=0)[0]
    display(Temp.style\
    .format({'coef': '{:.4e}', 'P>|t|': '{:.4e}', 'std err': '{:.4e}'})\
    .bar(subset=['coef'], align='mid', color='Lime')\
    .set_properties(subset=['std err'], **{'background-color': 'DimGray', 'color': 'White'}))

Reg_Result(Results)
```

The provided output is from the summary of a fitted linear regression model, which provides valuable information about the model's coefficients, their standard errors, t-values, p-values, and confidence intervals. Let's break down each component:

1. `const` and `LSTAT`: These are the variable names in the model. `const` represents the intercept (constant term), and `LSTAT` represents the predictor variable.

2. `coef`: This column displays the estimated coefficients of the linear regression model. For `const`, the estimated coefficient is 34.5538, and for `LSTAT`, it is -0.9500.

3. `std err`: This column represents the standard errors of the coefficient estimates. It quantifies the uncertainty in the estimated coefficients. For `const`, the standard error is 0.563, and for `LSTAT`, it is 0.039.

4. `t`: The t-values are obtained by dividing the coefficient estimates by their standard errors. The t-value measures how many standard errors the coefficient estimate is away from zero. For `const`, the t-value is 61.415, and for `LSTAT`, it is -24.528.

5. `P>|t|`: This column provides the p-values associated with the t-values. The p-value represents the probability of observing a t-value as extreme as the one calculated, assuming the null hypothesis that the coefficient is equal to zero. Lower p-values indicate stronger evidence against the null hypothesis. In this case, both coefficients have extremely low p-values (close to 0), indicating that they are statistically significant.

6. `[0.025 0.975]`: These are the lower and upper bounds of the 95% confidence intervals for the coefficients. The confidence intervals provide a range of plausible values for the true population coefficients. For `const`, the confidence interval is [33.448, 35.659], and for `LSTAT`, it is [-1.026, -0.874].

In summary, this output shows the results of fitting a simple linear regression model with `medv` as the response variable and `lstat` as the predictor variable. The coefficient estimates indicate how the response variable (`medv`) is expected to change for a one-unit increase in the predictor variable (`lstat`). The low p-values and confidence intervals not containing zero suggest that both the intercept and `LSTAT` coefficient are statistically significant and have a significant impact on the response variable.

In statsmodels, the `get_influence()` method provides valuable diagnostic tools for examining the influence of individual observations on a fitted linear regression model. When you fit a linear regression model using statsmodels, the `get_influence()` method can be called on the result object to access an instance of the `Influence` class. This `Influence` object contains various diagnostic measures that help you identify influential observations and potential outliers in your regression analysis.

Here's a more detailed explanation of the common diagnostic measures accessible from the `Influence` object:

1. Cook's distance (`cooks_distance`):
   Cook's distance is a measure of the influence of each observation on the fitted values of the regression model. It quantifies how much the model's predicted values change when a specific observation is omitted. Large Cook's distances indicate influential observations that significantly impact the regression model's results. These observations may have a considerable effect on the estimated coefficients and, if omitted, could substantially alter the model's conclusions.

2. Studentized residuals (`resid_studentized_external`):
   Studentized residuals are the residuals of the model divided by their estimated standard deviations. They are used to detect potential outliers in the data. Observations with large absolute values of studentized residuals may be influential points that do not conform to the assumed error distribution of the model. These observations might represent unusual or extreme cases that could distort the model's performance.

3. Hat matrix diagonal (`hat_matrix_diag`):
   The hat matrix is a matrix that projects the target variable onto the fitted values, taking into account the weights of each observation. The diagonal elements of the hat matrix represent the leverage of each observation. Leverage indicates how far an observation's predictor variable values deviate from the average values in the dataset. Observations with high leverage can have a significant impact on the regression coefficients.

4. DFBETAS (`dfbetas`):
   DFBETAS are the changes in the estimated regression coefficients when each observation is excluded from the model one at a time. They help identify the influence of individual observations on the model's parameter estimates. Large DFBETAS suggest that a particular observation has a substantial effect on the coefficients, and its inclusion significantly affects the model's results.

By examining these diagnostic measures, you can gain insights into the sensitivity of your regression model to individual observations and detect potential outliers or influential data points. Understanding and addressing influential observations are crucial in building robust and reliable regression models.

## Multiple Linear Regression

To perform multiple linear regression using least squares in Python, we can use the `ModelSpec()` function from the appropriate library, such as `statsmodels` or `scikit-learn`. The `ModelSpec()` function helps construct the model matrix and response variables necessary for the regression analysis. In this example, we will consider a multiple linear regression model with two predictor variables: `lstat` and `age`.

Here's how you can do it using `statsmodels`:

```python
import statsmodels.api as sm

# Assuming you have the data in a pandas DataFrame named 'data'
# where 'lstat' and 'age' are the predictor variables, and 'target' is the target variable.

# Step 1: Prepare the predictor variables and target variable
X = Boston[['LSTAT', 'AGE']]  # Predictor variables
y = Boston['MEDV']          # Target variable

# Step 2: Add a constant term to the predictor variables (intercept)
X = sm.add_constant(X)

# Step 3: Fit the multiple linear regression model using ordinary least squares (OLS)
model = sm.OLS(y, X)
Results = model.fit()

# Step 4: Print the summary of the regression results
print(Results.summary())
Reg_Result(Results)
```

5. Method: Least Squares

It assesses whether the model as a whole is statistically significant in explaining the variance in the dependent variable.    - Omnibus: The Omnibus test is a test for the normality of the residuals. In this case, the low probability (p-value < 0.001) suggests that the residuals are not normally distributed.

   - Prob(Omnibus): The probability associated with the Omnibus test.
   - Jarque-Bera (JB): The Jarque-Bera test is another test for the normality of the residuals. The low probability (p-value < 0.001) indicates that the residuals do not follow a normal distribution.
   - Skew: Skewness measures the asymmetry of the residuals' distribution. Positive skewness (as in this case) indicates a right-skewed distribution.
   - Kurtosis: Kurtosis measures the heaviness of the tails of the residuals' distribution. A value greater than 3 indicates heavier tails compared to the normal distribution.
   - Cond. No.: The condition number measures the multicollinearity among predictor variables. A high condition number (>30) suggests possible multicollinearity issues.

Overall, this summary provides a comprehensive evaluation of the multiple linear regression model's performance and the significance of each predictor variable in explaining the variation in the dependent variable.

In linear regression modeling, we have the flexibility to define our linear model in various ways, including adding interaction terms between predictor variables. In the provided example, we are considering a linear model that uses 'Radio', 'TV', and an interaction term between the two ('Radio' multiplied by 'TV') to predict 'Sales'. The model can be expressed in two equivalent forms:

1. **Additive Model Form**:
\begin{equation}Sales = \beta_0 + \beta_1 \times TV + \beta_2 \times \text{Radio} + \beta_3 \times (\text{Radio} \times \text{TV}) + \epsilon\end{equation}

2. **Multiplicative Model Form**:
\begin{equation}Sales = \beta_0 + (\beta_1 + \beta_3 \times \text{Radio}) \times \text{TV} + \beta_2 \times \text{Radio} + \epsilon\end{equation}

Both forms of the model yield the same predictions for 'Sales' based on the values of 'TV' and 'Radio'. The difference lies in how the model parameters ($\beta_0$, $\beta_1$, $\beta_2$, and $\beta_3$) are interpreted in each form:

- In the additive model form, $\beta_1$ represents the effect of 'TV' on 'Sales', $\beta_2$ represents the effect of 'Radio' on 'Sales', and $\beta_3$ represents the additional effect on 'Sales' due to the interaction between 'TV' and 'Radio'.

- In the multiplicative model form, $\beta_1 + \beta_3 \times \text{Radio}$ represents the combined effect of 'TV' and the interaction term ($\text{Radio} \times \text{TV}$) on 'Sales', while $\beta_2$ still represents the effect of 'Radio' on 'Sales'.

Both forms allow us to examine the contributions of 'TV', 'Radio', and their interaction to the prediction of 'Sales'. The decision to include interaction terms depends on the nature of the data and the research question at hand. Interaction terms are used when there is reason to believe that the combined effect of two predictor variables is not simply the sum of their individual effects.

In summary, both the additive and multiplicative forms of the linear model are valid representations of the same underlying relationship between the predictor variables and the outcome. The choice between the two forms depends on the specific context of the analysis and the hypotheses being tested.

When dealing with predictors that have only two levels (binary variables), we can create indicator or dummy variables to represent these levels numerically.

For example, to handle the predictor variable "Married," which has two levels (Married and not Married), we can create a binary indicator or dummy variable to represent these levels numerically.

Let's create a new binary variable, denoted as $x_i$, that takes the following form {cite:p}`james2023introduction`:

\begin{equation}
x_i = \begin{cases}
1, & \text{if the $i$th person is Married}, \\
0, & \text{if the $i$th person is not Married}.
\end{cases}
\end{equation}

The regression equation involving this binary variable would be:

\begin{equation}
y_i = \beta_0 + \beta_1 x_i + \epsilon_i = \begin{cases}
\beta_0 + \beta_1 + \epsilon_i, & \text{if the $i$th person is Married}, \\
\beta_0 + \epsilon_i, & \text{if the $i$th person is not Married}.
\end{cases}
\end{equation}

In this regression equation, $\beta_0$ represents the intercept, $\beta_1$ represents the coefficient associated with the binary variable $x_i$, and $\epsilon_i$ is the error term for the $i$th observation. When $x_i = 1$, indicating that the $i$th person is Married, the equation becomes $y_i = \beta_0 + \beta_1 + \epsilon_i$. On the other hand, when $x_i = 0$, indicating that the $i$th person is not Married, the equation becomes $y_i = \beta_0 + \epsilon_i$ {cite:p}`james2023introduction`.

By using binary variables for predictors with two levels, we can incorporate categorical information into regression models and study how the presence or absence of a particular category influences the response variable. This approach allows us to handle categorical predictors effectively in linear regression models.

When dealing with qualitative predictors that have more than two levels, a single dummy variable cannot represent all possible values. In this situation, we can create additional dummy variables, one for each level of the qualitative predictor. Let's take the "region" variable as an example, which has three levels: South, West, and East.

We create two dummy variables as follows {cite:p}`james2023introduction`:

1. $x_{i1}$:

   - $x_{i1} = 0$ if the $i$th person is not from the South

2. $x_{i2}$:

   - $x_{i2} = 0$ if the $i$th person is not from the West

Then both of these variables can be used in the regression equation to obtain the model:

\begin{equation}
y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \epsilon_i =
\begin{cases}
\beta_0 + \beta_1 + \epsilon_i, & \text{if the $i$th person is from the South}, \\
\beta_0 + \beta_2 + \epsilon_i, & \text{if the $i$th person is from the West}, \\

In general, when we have a qualitative predictor with $k$ levels, we create $k-1$ dummy variables. The level with no dummy variable (in this case, "East") is known as the baseline, against which the other levels are compared. This approach allows us to include qualitative predictors with multiple levels in regression models and understand how each level influences the response variable in comparison to the baseline.

<font color='Blue'><b>Example</b></font>: In the previous example, we considered the concept of interactions in the context of the Credit dataset. We wanted to predict the credit card balance (response variable) using income (quantitative) and student status (qualitative) variables.

Without an interaction term, the model takes the form:

\begin{align}
\text{balance}_i & \approx \beta_0 + \beta_1 \times \text{income}_i + \begin{cases}
\beta_2, & \text{if the $i$th person is a student}, \\
0, & \text{if the $i$th person is not a student}.
\end{cases}
\notag \\
& = \begin{cases}
\beta_1 \times \text{income}_i + (\beta_0 + \beta_2), & \text{if the $i$th person is a student}, \\
\beta_1 \times \text{income}_i + \beta_0, & \text{if the $i$th person is not a student}.
\end{cases}
\end{align}

As shown in the left-hand panel of Figure 3.7, this model fits two parallel lines to the data, one for students and one for non-students. The lines have different intercepts, $\beta_0 + \beta_2$ versus $\beta_0$, but the same slope, $\beta_1$. This implies that the average effect on credit card balance of a one-unit increase in income does not depend on whether or not the individual is a student. However, in reality, a change in income may have a different effect on the credit card balance of a student versus a non-student.

To address this limitation, we add an interaction variable by multiplying income with the dummy variable for student. The new model becomes:

\begin{align}
\text{balance}_i & \approx \beta_0 + \beta_1 \times \text{income}_i + \begin{cases}
\beta_2 + \beta_3 \times \text{income}_i, & \text{if the $i$th person is a student}, 
\\
0, & \text{if the $i$th person is not a student}.
\end{cases}
\notag \\
& = \begin{cases}
(\beta_0 + \beta_2) + (\beta_1 + \beta_3) \times \text{income}_i, & \text{if the $i$th person is a student}, \\
\beta_0 + \beta_1 \times \text{income}_i, & \text{if the $i$th person is not a student}.
\end{cases}
\end{align}

As illustrated in the right-hand panel of Figure 3.7, this model also fits two regression lines for students and non-students. The lines have different intercepts, $\beta_0 + \beta_2$ versus $\beta_0$, as well as different slopes, $\beta_1 + \beta_3$ versus $\beta_1$. This allows for the possibility that changes in income may affect the credit card balances of students and non-students differently.

Adding an interaction term enables the model to capture varying effects of income on credit card balances based on student status, making it more flexible and accurate in representing the data. The interaction term allows us to account for potential differences in the relationship between income and credit card balance between students and non-students.

**Regression 1 - without interaction term:**

```python
Results2 = smf.ols('Balance ~ Income + Income*Student', Credit).fit()
reg2 = Results2.params
print(Results2.summary())
Reg_Result(Results2)
```
